## pytorch中的学习率调整策略

    class _LRScheduler(object)
    
**基本属性**

- optimizer

    关联的优化器
    
- last_epoch

    记录epoch数
    
- base_lrs

    记录初始学习率

**基本方法**

- step()

    更新下一个epoch的学习率
    
- get_lr()

    计算下一个epoch的学习率 

**StepLR方法**

- 功能

    等间隔调整学习率
    
- step_size

    调整间隔数
    
- gamma
    
    调整系数
    
- 调整公式

    $$l_r=l_r*gamma$$

```python
%matplotlib inline
import torch
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
torch.manual_seed(1)


'''
数据准备 
'''
LR = 0.1   #初始学习率
iteration = 10
max_epoch = 200
# ------------------------------ fake data and optimizer  ------------------------------

weights = torch.randn((1), requires_grad=True)
target = torch.zeros((1))

##设置优化器
optimizer = optim.SGD([weights], lr=LR, momentum=0.9)
```

```python
# ------------------------------ 1 Step LR ------------------------------

#载入关联优化器
scheduler_lr = optim.lr_scheduler.StepLR(optimizer, step_size=50, gamma=0.1)  # 设置学习率下降策略

lr_list, epoch_list = list(), list()
for epoch in range(max_epoch):

    lr_list.append(scheduler_lr.get_lr())  ##获取当前学习率
    epoch_list.append(epoch)

    for i in range(iteration):

        loss = torch.pow((weights - target), 2)
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()

    scheduler_lr.step()  ##在迭代轮数的循环中更新学习率

plt.plot(epoch_list, lr_list, label="Step LR Scheduler")
plt.xlabel("Epoch")
plt.ylabel("Learning rate")
plt.legend()
plt.show()
```

**MultiStepLR方法**

- 功能

    多间隔调整学习率
    
- milestones
    
    根据给定的list，设定调整时刻
    
- gamma
    
    调整系数

```python
# ------------------------------ 2 Multi Step LR ------------------------------

##时刻列表
milestones = [50, 125, 160]

##设置策略
scheduler_lr = optim.lr_scheduler.MultiStepLR(optimizer, milestones=milestones, gamma=0.1)

lr_list, epoch_list = list(), list()
for epoch in range(max_epoch):

    lr_list.append(scheduler_lr.get_lr())
    epoch_list.append(epoch)

    for i in range(iteration):

        loss = torch.pow((weights - target), 2)
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()

    scheduler_lr.step()

plt.plot(epoch_list, lr_list, label="Multi Step LR Scheduler\nmilestones:{}".format(milestones))
plt.xlabel("Epoch")
plt.ylabel("Learning rate")
plt.legend()
plt.show()
```

**ExponentialLR方法**

- 功能

    按指数衰减调整学习率
    
- gamma

    指数的底,通常接近于1的数

- 计算公式

    $$l_r=l_r*gamma^{epoch}$$

```python
# ------------------------------ 3 Exponential LR ------------------------------

gamma = 0.95
scheduler_lr = optim.lr_scheduler.ExponentialLR(optimizer, gamma=gamma)

lr_list, epoch_list = list(), list()
for epoch in range(max_epoch):

    lr_list.append(scheduler_lr.get_lr())
    epoch_list.append(epoch)

    for i in range(iteration):

        loss = torch.pow((weights - target), 2)
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()

    scheduler_lr.step()

plt.plot(epoch_list, lr_list, label="Exponential LR Scheduler\ngamma:{}".format(gamma))
plt.xlabel("Epoch")
plt.ylabel("Learning rate")
plt.legend()
plt.show()
```

**CosineAnnealingLR**

- 功能

    余弦周期调整学习率
    
- T_max

    下降周期
    
- eta_min

    学习率下限
    
- 调整方式(1+cos(\frac T_{cur}T_{max})

    $$\eta_t=\eta_{min}+\frac 12 *(\eta_{max}-\eta_{min})*(1+cos(\frac {T_{cur}}{T_{max}})\pi)$$


```python
# ------------------------------ 4 Cosine Annealing LR ------------------------------

t_max = 50
scheduler_lr = optim.lr_scheduler.CosineAnnealingLR(optimizer,
                                                    T_max=t_max,
                                                    eta_min=0.)

lr_list, epoch_list = list(), list()
for epoch in range(max_epoch):

    lr_list.append(scheduler_lr.get_lr())
    epoch_list.append(epoch)

    for i in range(iteration):

        loss = torch.pow((weights - target), 2)
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()

    scheduler_lr.step()

plt.plot(epoch_list,
         lr_list,
         label="CosineAnnealingLR Scheduler\nT_max:{}".format(t_max))
plt.xlabel("Epoch")
plt.ylabel("Learning rate")
plt.legend()
plt.show()
```

**ReduceLRonPlateau方法**

- 功能

    监控指标当指标不再变化时则调整
    
- mode

    min/max 两种模式
    
    - loss函数使用min模式
    - 准确率通常使用max模式
    
- factor

    调整系数
    
- patience

    耐心值，接受几次不变化
    
- cooldown

    冷却时间，停止观察一段时间
    
- verbose

    是否打印日志
    
- min_lr

    学习率下限
    
- eps

    学习率衰减最小值

```python
# ------------------------------ 5 Reduce LR On Plateau ------------------------------

loss_value = 0.5
accuray = 0.9

factor = 0.1
mode = "min"
patience = 10
cooldown = 10
min_lr = 1e-4
verbose = True

scheduler_lr = optim.lr_scheduler.ReduceLROnPlateau(optimizer,
                                                    factor=factor,
                                                    mode=mode,
                                                    patience=patience,
                                                    cooldown=cooldown,
                                                    min_lr=min_lr,
                                                    verbose=verbose)

for epoch in range(max_epoch):
    for i in range(iteration):

        # train(...)

        optimizer.step()
        optimizer.zero_grad()

    if epoch == 5:
        loss_value = 0.4

    scheduler_lr.step(loss_value)  ##注意把衡量指标传入学习率调整参数
```

**LambdaLR方法**

- 功能

    自定义调整策略

- lr_lambda

    自定义函数或者list

```python
# ------------------------------ 6 lambda ------------------------------

lr_init = 0.1

weights_1 = torch.randn((6, 3, 5, 5))
weights_2 = torch.ones((5, 5))

optimizer = optim.SGD([{
    'params': [weights_1]
}, {
    'params': [weights_2]
}],
                      lr=lr_init)

##自定义学习率更新函数
lambda1 = lambda epoch: 0.1**(epoch // 20)
lambda2 = lambda epoch: 0.95**epoch

scheduler = torch.optim.lr_scheduler.LambdaLR(optimizer,
                                              lr_lambda=[lambda1, lambda2])

lr_list, epoch_list = list(), list()
for epoch in range(max_epoch):
    for i in range(iteration):

        # train(...)

        optimizer.step()
        optimizer.zero_grad()

    scheduler.step()

    lr_list.append(scheduler.get_lr())
    epoch_list.append(epoch)

    print('epoch:{:5d}, lr:{}'.format(epoch, scheduler.get_lr()))

plt.plot(epoch_list, [i[0] for i in lr_list], label="lambda 1")
plt.plot(epoch_list, [i[1] for i in lr_list], label="lambda 2")
plt.xlabel("Epoch")
plt.ylabel("Learning Rate")
plt.title("LambdaLR")
plt.legend()
plt.show()
```
